Notebook para visualizar a evolução de uma estratégia de aposta a longo prazo

```python
import random
import numpy as np


def build_list(params, num):
    result = [value for value, prob in params for _ in range(int(prob * num))]
    random.shuffle(result)
    return result
```

```python
taxa_acerto = 0.94

premio = 1
prejuizo = -7

num_chances = 1000

num_possibilidades = 12
possibilidades = []
```

```python
for p in range(0,num_possibilidades):
    
    random_list = build_list([(1,taxa_acerto),(-1,(1-taxa_acerto))], num_chances)
    random_list = [premio if rl == 1 else prejuizo for rl in random_list]
    random_list = np.cumsum(random_list)
    
    possibilidades.append(random_list)
    
```

```python
import matplotlib.pyplot as plt

for impacts in possibilidades:
     t = plt.plot(impacts)

plt.show()
```
